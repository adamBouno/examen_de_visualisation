## 0. Bloc commun — Importation des librairies et fonctions utilitaires

On importe toutes les librairies nécessaires et on définit les fonctions et constantes partagées par l'ensemble du notebook.

````python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import f1_score, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from imblearn.over_sampling import RandomOverSampler

sns.set_theme(style="whitegrid")
%matplotlib inline

# Création des dossiers de sortie
Path("sorties_exam/figures").mkdir(parents=True, exist_ok=True)
Path("sorties_exam/datasets_equilibres").mkdir(parents=True, exist_ok=True)

DATASETS = [
    "abalone9-18.csv",
    "ecoli-0-1-4-7_vs_5-6.csv",
    "ecoli4.csv",
    "glass-0-1-6_vs_2.csv",
    "glass-0-1-6_vs_5.csv",
    "glass2.csv",
    "glass4.csv",
    "page-blocks-1-3_vs_4.csv",
    "shuttle-c0-vs-c4.csv",
    "yeast-1_vs_7.csv",
]

MODELS = {
    "LogReg":       LogisticRegression(max_iter=2000),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(n_estimators=80, random_state=42),
    "KNN":          KNeighborsClassifier(),
    "NaiveBayes":   GaussianNB(),
}
````

````python
def load_dataset(file):
    """Charge un CSV, nettoie les noms de colonnes et détecte la colonne cible."""
    df = pd.read_csv(file, skipinitialspace=True)
    df.columns = [c.strip().replace("{", "").strip() for c in df.columns]
    class_col = next(c for c in df.columns if c.lower() == "class")
    df[class_col] = df[class_col].astype(str).str.strip()
    return df, class_col

def numeric_cols(df, class_col):
    """Retourne la liste des colonnes numériques (hors cible)."""
    return [c for c in df.columns if c != class_col]
````

## Question 1 — Charger les 10 datasets déséquilibrés et visualiser les classes

On charge les 10 datasets, on résume les effectifs par classe, puis on visualise la distribution avec **Seaborn** (`barplot`).

**Résumé des effectifs :**

| Dataset | Classe majoritaire | Classe minoritaire |
|---|---:|---:|
| abalone9-18 | negative: 689 | positive: 42 |
| ecoli-0-1-4-7_vs_5-6 | negative: 307 | positive: 25 |
| ecoli4 | negative: 316 | positive: 20 |
| glass-0-1-6_vs_2 | negative: 175 | positive: 17 |
| glass-0-1-6_vs_5 | negative: 175 | positive: 9 |
| glass2 | negative: 197 | positive: 17 |
| glass4 | negative: 201 | positive: 13 |
| page-blocks-1-3_vs_4 | negative: 444 | positive: 28 |
| shuttle-c0-vs-c4 | negative: 1706 | positive: 123 |
| yeast-1_vs_7 | negative: 429 | positive: 30 |

````python
rows = []
for f in DATASETS:
    df, c = load_dataset(f)
    for k, v in df[c].value_counts().items():
        rows.append({"dataset": Path(f).stem, "classe": k, "effectif": v})

summary = pd.DataFrame(rows)
print(summary.to_string(index=False))
````

````python
plt.figure(figsize=(16, 5))
ax = sns.barplot(data=summary, x="dataset", y="effectif", hue="classe", palette="Set2")
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
plt.title("Distribution des classes — 10 datasets déséquilibrés (Seaborn)")
plt.tight_layout()
plt.savefig("sorties_exam/figures/q1_original_classes_seaborn.png", dpi=150)
plt.show()
````

## Question 2 — Équilibrer les datasets avec Weka et enregistrer en CSV

### Procédure Weka (interface graphique)

1. Ouvrir **Weka Explorer**.
2. Aller dans **Preprocess** puis **Open file**.
3. Charger un fichier CSV, par exemple `yeast-1_vs_7.csv`.
4. Vérifier que l'attribut `Class` ou `class` est bien le dernier attribut.
5. Cliquer sur **Choose** dans la zone **Filter**.
6. Choisir `supervised > instance > SMOTE`.
7. Régler par exemple `percentage = 1300` pour augmenter la classe minoritaire, ou adapter selon le taux de déséquilibre.
8. Cliquer sur **Apply**.
9. Vérifier les nouveaux effectifs de classes.
10. Cliquer sur **Save** et enregistrer avec le nom `mon_dataset_nom.csv`.
11. Répéter pour les 10 datasets.

> **Alternative Weka :** utiliser `supervised > instance > Resample` avec `biasToUniformClass = 1.0`.

### Reproduction en Python (équivalent SMOTE via `RandomOverSampler`)

Pour tester la suite sans manipulation manuelle dans Weka, on génère les fichiers équilibrés dans `sorties_exam/datasets_equilibres/`.

````python
for f in DATASETS:
    df, c = load_dataset(f)
    X, y = df[numeric_cols(df, c)], df[c]
    X_res, y_res = RandomOverSampler(random_state=42).fit_resample(X, y)
    df_res = pd.DataFrame(X_res, columns=X.columns)
    df_res[c] = y_res
    out = f"sorties_exam/datasets_equilibres/mon_dataset_{Path(f).stem}.csv"
    df_res.to_csv(out, index=False)
    print(f"{Path(out).name:50s} -> {df_res[c].value_counts().to_dict()}")
````

## Question 3 — Re-visualiser les classes équilibrées

On recharge les versions équilibrées et on affiche la nouvelle distribution des classes. Chaque dataset est désormais parfaitement équilibré (50% / 50%).

````python
balanced_files = sorted(Path("sorties_exam/datasets_equilibres").glob("mon_dataset_*.csv"))
rows_bal = []
for f in balanced_files:
    df, c = load_dataset(f)
    for k, v in df[c].value_counts().items():
        rows_bal.append({"dataset": f.stem, "classe": k, "effectif": v})

balanced_summary = pd.DataFrame(rows_bal)
````

````python
# Seaborn
plt.figure(figsize=(16, 5))
ax = sns.barplot(data=balanced_summary, x="dataset", y="effectif", hue="classe", palette="Set1")
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")
plt.title("Distribution des classes — 10 datasets équilibrés (Seaborn)")
plt.tight_layout()
plt.show()
````

````python
# Plotly Express
fig = px.bar(
    balanced_summary, x="dataset", y="effectif",
    color="classe", barmode="group",
    title="Distribution des classes — 10 datasets équilibrés (Plotly)"
)
fig.write_html("sorties_exam/figures/q3_equilibre_classes_plotly.html")
fig.show()
````

## Question 4 — Cinq types de visualisation pour chaque dataset original

Types choisis pour chaque dataset :
1. **Diagramme de classes** (`countplot`) — répartition des classes.
2. **Histogramme** — distribution d'une variable numérique, colorée par classe.
3. **Nuage de points** — relation entre deux variables.
4. **Boîte à moustaches** — distribution d'une variable par classe, détection d'outliers.
5. **Matrice de corrélation** (`heatmap`) — corrélations entre variables numériques.

````python
for f in DATASETS:
    df, c = load_dataset(f)
    nums = numeric_cols(df, c)
    title = Path(f).stem

    fig, axes = plt.subplots(1, 5, figsize=(26, 4))

    # 1. Diagramme de classes
    sns.countplot(data=df, x=c, ax=axes[0], palette="Set2",
                  hue=c, legend=False)
    axes[0].set_title("Répartition des classes")

    # 2. Histogramme
    sns.histplot(data=df, x=nums[0], hue=c, kde=True, ax=axes[1])
    axes[1].set_title(f"Histogramme ({nums[0]})")

    # 3. Nuage de points
    sns.scatterplot(data=df, x=nums[0], y=nums[1], hue=c, ax=axes[2], alpha=0.7)
    axes[2].set_title(f"Scatter ({nums[0]} vs {nums[1]})")

    # 4. Boîte à moustaches
    sns.boxplot(data=df, x=c, y=nums[0], ax=axes[3], palette="pastel",
                hue=c, legend=False)
    axes[3].set_title(f"Boxplot ({nums[0]})")

    # 5. Heatmap de corrélation
    corr_cols = nums[:8]
    sns.heatmap(df[corr_cols].corr(), cmap="vlag", center=0,
                ax=axes[4], annot=False)
    axes[4].set_title("Heatmap corrélation")

    plt.suptitle(title, fontsize=13, fontweight="bold")
    plt.tight_layout()
    plt.show()
````

## Question 6 — F1-Measure avec cinq algorithmes et histogramme

Dataset choisi : **yeast-1_vs_7.csv** (déséquilibré vs équilibré).  
Algorithmes : Logistic Regression, Decision Tree, Random Forest, KNN, Naive Bayes.

**Résultats obtenus :**

| Algorithme | F1 original | F1 équilibré |
|---|---:|---:|
| Logistic Regression | 0.000 | 0.304 |
| Decision Tree | 0.556 | 0.308 |
| Random Forest | 0.333 | 0.364 |
| KNN | 0.200 | 0.188 |
| Naive Bayes | 0.158 | 0.155 |

**Interprétation :** l'équilibrage améliore le F1 pour les modèles linéaires (LogReg) et les forêts aléatoires, mais l'effet varie selon l'algorithme.

````python
df_y, c_y = load_dataset("yeast-1_vs_7.csv")
X_y, y_y = df_y[numeric_cols(df_y, c_y)], df_y[c_y]

Xtr, Xte, ytr, yte = train_test_split(
    X_y, y_y, test_size=0.3, stratify=y_y, random_state=42
)
Xb, yb = RandomOverSampler(random_state=42).fit_resample(Xtr, ytr)

rows_f1 = []
for name, model in MODELS.items():
    for version, Xfit, yfit in [("original", Xtr, ytr), ("equilibre", Xb, yb)]:
        model.fit(Xfit, yfit)
        score = f1_score(yte, model.predict(Xte), pos_label="positive")
        rows_f1.append({"algo": name, "version": version, "f1": score})

f1_df = pd.DataFrame(rows_f1)
print(f1_df.pivot(index="algo", columns="version", values="f1").round(3))
````

````python
fig = px.histogram(
    f1_df, x="algo", y="f1", color="version",
    barmode="group", histfunc="sum",
    title="F1-Measure — 5 algorithmes, yeast-1_vs_7 (original vs équilibré)",
    labels={"algo": "Algorithme", "f1": "F1-Measure", "version": "Version"}
)
fig.write_html("sorties_exam/figures/q6_f1.html")
fig.show()
````

## Question 7 — Accuracy avec cinq algorithmes et diagramme en boîte

Validation croisée 5-fold sur **yeast-1_vs_7.csv** (original et équilibré).

**Moyennes d'accuracy obtenues :**

| Algorithme | Original | Équilibré |
|---|---:|---:|
| Decision Tree | 0.887 | 0.960 |
| KNN | 0.935 | 0.922 |
| Logistic Regression | 0.935 | 0.767 |
| Naive Bayes | 0.264 | 0.575 |
| Random Forest | 0.939 | 0.992 |

````python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

df_bal_y, c_bal_y = load_dataset(
    "sorties_exam/datasets_equilibres/mon_dataset_yeast-1_vs_7.csv"
)

rows_acc = []
for version, data_v, target_v in [
    ("original", df_y, c_y),
    ("equilibre", df_bal_y, c_bal_y)
]:
    X_v = data_v[numeric_cols(data_v, target_v)]
    y_v = data_v[target_v]
    for name, model in MODELS.items():
        for score in cross_val_score(model, X_v, y_v, cv=cv, scoring="accuracy"):
            rows_acc.append({"algo": name, "version": version, "accuracy": score})

acc_df = pd.DataFrame(rows_acc)
print(acc_df.groupby(["algo", "version"])["accuracy"].mean().unstack().round(3))
````

````python
fig = px.box(
    acc_df, x="algo", y="accuracy", color="version",
    title="Accuracy (CV 5-fold) — 5 algorithmes, yeast-1_vs_7",
    labels={"algo": "Algorithme", "accuracy": "Accuracy", "version": "Version"}
)
fig.write_html("sorties_exam/figures/q7_accuracy_boxplot.html")
fig.show()
````

## Question 8 — Matrice de confusion

Algorithme choisi : **Random Forest**.  
Dataset : **yeast-1_vs_7.csv** — version originale et version équilibrée.

````python
clf = RandomForestClassifier(n_estimators=80, random_state=42)

# Version originale
clf.fit(Xtr, ytr)
fig_cm, ax_cm = plt.subplots(figsize=(5, 4))
ConfusionMatrixDisplay.from_estimator(clf, Xte, yte, ax=ax_cm, colorbar=False)
ax_cm.set_title("Matrice de confusion — Random Forest (original)")
plt.tight_layout()
plt.savefig("sorties_exam/figures/q8_confusion_original.png", dpi=150)
plt.show()

# Version équilibrée
clf.fit(Xb, yb)
fig_cm2, ax_cm2 = plt.subplots(figsize=(5, 4))
ConfusionMatrixDisplay.from_estimator(clf, Xte, yte, ax=ax_cm2, colorbar=False)
ax_cm2.set_title("Matrice de confusion — Random Forest (équilibré)")
plt.tight_layout()
plt.savefig("sorties_exam/figures/q8_confusion_equilibre.png", dpi=150)
plt.show()
````
